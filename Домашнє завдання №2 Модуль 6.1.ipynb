Домашнє завдання №2

1. В урні знаходиться 15 білих, 5 червоних і 10 чорних куль. Випадковим чином витягується 1 куля, знайти ймовірність того, що вона буде: а) білою, б) червоною, в) чорною.

P(біла) = 15/30 = 1/2

P(червона) = 5/30 = 1/6

P(чорна) = 10/30 = 1/3

2. В магазин надійшло 30 холодильників, п'ять з яких мають заводський дефект. Випадковим чином обирають один холодильник. Яка ймовірність того, що він буде без дефекту?

Кількість холодильників без дефекту = 30-5 = 25
P(холодильник без дефекту) = 25/30 = 5/6

3. Набираючи номер телефону, абонент забув дві останні цифри, але пам'ятає, що одна з них — нуль, а інша — непарна. Знайти ймовірність того, що він набере правильний номер.

У нас є два місця для двох цифр 0 і набір з непарних цифр яких п'ять: (1, 3, 5, 7, 9), оскільки порядок має значення, максимальна кількість варіантів 10, вірний варіант лише буде одним, тому відповідь буде P(набору правильного номеру) = 1/10

4. Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
    
    а) дорівнюватиме семи;
    б) буде не меншим за 20;
    в) буде парним.

а) оскільки у двох кубиках ми маємо два набори цифр (1, 2, 3, 4, 5, 6) і добуток таких цифр не може дорівнювати семи, то
P(добуток чисел дорівнюватиме семи) = 0;
б) Якщо перебрати усі можливі добутки двох наборів цифр, які >= 20, в нас вийде 8 комбінацій, максимальна кількість комбінацій 36, відповідно
P(добуток чисел, яких буде не меншим за 20) = 8/36 = 2/9
в) Якщо перебрати усі можливі добутки двох наборів цифр, парних добутків вийде 27, відповідно
P (добуток чисел буде парним) = 27/36 = 3/4/

5. Ймовірність того, що на тесті з біології учень О. правильно розв'яже більше 11 задач, дорівнює 0,67. Ймовірність того, що О. правильно розв'яже більше 10 задач, дорівнює 0,74. Знайдіть ймовірність того, що О. правильно розв'яже рівно 11 задач.

P(A > 10) — це ймовірність, що включає P(A = 11) і P(A > 11), віднімаючи P(A > 11) від P(A > 10), ми виключаємо всі випадки, коли P(A > 11), і залишаємо лише випадок коли P(A=11) = 0.74 - 0.67 = 0.07
